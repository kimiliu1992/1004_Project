```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

taxi_o = pd.read_csv('Taxi_Data.csv',index_col=0)
taxi_0 = taxi_o[taxi_o.Location_ID ==0]
taxi = taxi_o[taxi_o.Location_ID !=0]
uber2014 = pd.read_csv('Uber-2014.csv',index_col=0)
uber2015 = pd.read_csv('Uber-2015.csv',index_col=0)
```

```python
uber2014['Mon_Counts'] = uber2014.groupby(['Month'])['date'].transform('count')
uber2015['Mon_Counts'] = uber2015.groupby(['Month'])['Date'].transform('count')

uber2014['Loc_Counts'] = uber2014.groupby(['Location_ID'])['Month'].transform('count')
uber2015['Loc_Counts'] = uber2015.groupby(['locationID'])['Month'].transform('count')

uber2014 = uber2014.sort(columns='Month', axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
uber2015 = uber2015.sort(columns='Month', axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')

#taxi['Loc_Counts'] = taxi.groupby(['Location_ID'])['Month'].transform('count')
#taxi['Mon_Counts'] = taxi.groupby(['Month'])['Date'].transform('count')
```

# Monthly Counts Uber VS Taxi

```python
#uber 2014
a = uber2014.Mon_Counts.unique().tolist()
b = sorted(uber2014.Month.unique().tolist())
u = ['Uber_Counts']

df2014 = pd.DataFrame(a,b,columns=u)
df2014.plot(kind='bar',width=0.35,rot=0, color = 'skyblue', alpha= 0.7, edgecolor='w')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Monthly Trip Counts', fontsize=14)
plt.title('Uber 2014 Apr to Sept', fontsize = 14)
plt.show()
```

```python
#uber2015
c = uber2015.Mon_Counts.unique().tolist()
d = sorted(uber2015.Month.unique().tolist())
df2015 = pd.DataFrame(c,d,columns=u)
df2015.plot(kind='bar',width=0.35,rot=0, color = 'purple', alpha= 0.7, edgecolor='w')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Monthly Trip Counts', fontsize=14)
plt.title('Uber 2015 Jan to Jun', fontsize = 14)
plt.show()
```

```python
#taxi
taxi2014 = taxi[taxi['Year'] == 2014.]
taxi2015 = taxi[taxi['Year'] == 2015.]
mon1 = [4,5,6,7,8,9]
mon2 = [1,2,3,4,5,6]
taxi2014_49 = taxi2014.loc[taxi2014['Month'].isin(mon1)]
taxi2015_16 = taxi2015.loc[taxi2015['Month'].isin(mon2)]
```

```python
#taxi 2014
tm1 = taxi2014_49.groupby(['Month'])['Count'].sum().tolist()
#tm1 = taxi2014_49.Mon_Counts.unique().tolist()
t = ['Taxi_Counts']
dftm1 = pd.DataFrame(tm1, mon1,columns=t)
dftm1.plot(kind='bar',width=0.35,rot=0, color = 'orange', alpha= 0.7, edgecolor='w')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Monthly Trip Counts', fontsize=14)
plt.title('Taxi 2014 Apr to Sept', fontsize = 14)
plt.show()
```

# Locking location of big demand of Uber

```python
zhu = uber2014.groupby(['Location_ID'])['Mon_Counts'].count()
zhu = pd.DataFrame(zhu)
zhu = zhu.sort(columns= 'Mon_Counts', ascending=False)
zhu = zhu[zhu.Mon_Counts >50000]
plt.figure(dpi=1000, frameon=True,figsize=(10,4))
zhu.plot(kind='bar',width=0.8,rot=0, color = 'purple', alpha= 0.5, edgecolor='g')
```

```python
www = uber2014.groupby(['time','Location_ID'])['Loc_Counts'].count()
www = www.loc[16,:]
www = pd.DataFrame(www)
#www = www['Loc_Counts']
#www = pd.DataFrame(www)
www = www.sort(columns= 'Loc_Counts', ascending=False)
www = www[0:34]
www.plot(kind='barh',width=0.8,rot=0, color = 'lightgrey', alpha= 0.5, edgecolor='r')
```

```python
hers = uber2015.groupby(['time','locationID'])['Loc_Counts'].count()
#hers = pd.DataFrame(hers)
hers_peak = hers.loc[16, :]
#hers_peak = hers_peak['Loc_Counts']
hersss = pd.DataFrame(hers_peak)
hersss = hersss.sort(columns='Loc_Counts', ascending=False)
hersss = hersss[0:34]
plt.figure(dpi=1000, frameon=True,figsize=(12,4))
hersss.plot(kind='barh',width=0.8,rot=0, color = 'lightgreen', alpha= 0.5, edgecolor='g')
```

```python
ri = taxi2014_49.groupby(['Time','Location_ID'])['Count'].sum()
ri = ri.loc[16, :]
ri = pd.DataFrame(ri)
ri = ri.sort(columns='Count',ascending=False)
ri = ri[0:34]
plt.figure(dpi=1000, frameon=True,figsize=(10,12))
ri.plot(kind='barh',width=0.8,rot=0, color = 'y', alpha= 0.5, edgecolor='g')
```

```python
rio = taxi2015_16.groupby(['Time','Location_ID'])['Count'].sum()
rio = rio.loc[16, :]
rio = pd.DataFrame(rio)
rio = rio.sort(columns='Count',ascending=False)
rio = rio[0:34]
rio.plot(kind='barh',width=0.8,rot=0, color = 'r', alpha= 0.5, edgecolor='g')
```
