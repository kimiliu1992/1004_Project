```python
haha[haha['Location_ID_new'] == 1]['Count']
```

```python
ID_count = np.zeros(195)
```

```python
for i in range(195):
    if i+1 in haha['Location_ID_new'].tolist():
        ID_count[i] = haha[haha['Location_ID_new'] == i+1]['Count']
    else:
        ID_count[i] = 0
        
```

```python
type(ID_count)
```

```python
ids= ID_count.tolist()
ids
```

```python
ids = pd.DataFrame(ids, index= None)
```

```python
x = []
for i in xrange(1,195):
    x.append(i)
x = pd.DataFrame(x)
```

```python
result = pd.concat([x, ids], axis=1, ignore_index=True)
```
